```python
dimap={'M':'Malignant','B':'Begnin'}
```

```python
df['diagnosis']=df['diagnosis'].map(dimap)
```

```python
df['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
```

```python
to_drop=['Unnamed: 32','diagnosis','id']

```

```python
X=df.drop(to_drop,axis=1)
y=df['diagnosis']
```

```python
from sklearn.preprocessing import LabelEncoder 
y=LabelEncoder().fit_transform(y)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler,Normalizer
from sklearn.pipeline import make_pipeline
preproc=make_pipeline(StandardScaler(),Normalizer())
preproc.fit(X_train,y_train)
preproc.transform(X_train)
preproc.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier(n_estimators=10)
```

```python
rfc.fit(X_train,y_train)
```

```python
varimp=pd.DataFrame(rfc.feature_importances_,X.columns,columns=['Importance']).sort_values('Importance',ascending=False)
```

```python
varimp
```

```python
pred=rfc.predict(X_test)
```

```python
from sklearn import metrics
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print('classification report:', metrics.classification_report(y_test,pred))
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg=LogisticRegression()
```

```python
logreg.fit(X_train,y_train)
```

```python
pred=logreg.predict(X_test)
```

```python
print('classification report:' '\n' '\n', metrics.classification_report(y_test,pred))
print('\n')
sns.heatmap(metrics.confusion_matrix(y_test,pred),annot=True,cmap='coolwarm')
```

```python
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(X_train,y_train)
pred=nb.predict(X_test)
```

```python
print('Classification report:' '\n' '\n',metrics.classification_report(y_test,pred))
```

```python
sns.heatmap(metrics.confusion_matrix(y_test,pred),annot=True,cmap='coolwarm')
```

```python
from sklearn.tree import DecisionTreeClassifier
tree=DecisionTreeClassifier()
tree.fit(X_train,y_train)
pred=tree.predict(X_test)
```

```python
print('Classification report:' '\n' '\n',metrics.classification_report(y_test,pred))
```

```python
from sklearn.svm import SVC
svm=SVC(kernel='linear')
svm.fit(X_train,y_train)
pred=svm.predict(X_test)
```

```python
print('Classification report:' '\n' '\n',metrics.classification_report(y_test,pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5,metric = 'minkowski', p = 2)
knn.fit(X_train,y_train)
pred=knn.predict(X_test)
```

```python
print('Classification report:' '\n' '\n',metrics.classification_report(y_test,pred))
```
